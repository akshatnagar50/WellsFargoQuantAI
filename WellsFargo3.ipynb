```python
#for i in range(19,len(df),19):
data=k.loc[df['Tenor']=='2M']
data
```

```python
data.rename(columns = {'1':'1.0'}, inplace = True)
```

```python
k
```

```python
l='l'
lst=[]
for i in range(1,17):
  p=str(i/10)
  q=str(int(i+1)/10)
  r=str(int(i+2)/10)
  temp=data.loc[:,[p,q,r]]
  lst.append(temp)
```

```python
model=tf.keras.Sequential([tf.keras.layers.LSTM(512,input_shape=(300,1)),
                           tf.keras.layers.Dense(128),
                           tf.keras.layers.Dense(64),
                           tf.keras.layers.Dense(32),
                           tf.keras.layers.Dense(30)])
model.summary()
model.compile(loss='mean_squared_error',optimizer='Adam',metrics=['mean_absolute_error'])
```

```python
m=lst[0]
mns=[]
for i in range(0,11742,190):
  L=m.loc[i:i+(99*19)].to_numpy().flatten()
  M=m.loc[i+100*19:i+109*19].to_numpy().flatten()
  mns.append([L,M])
mns=np.array(mns)
mns=mns[0:60]
for i in range(0,50):
  model.fit(np.array([mns[i][0]]),np.array([mns[i][1]]),epochs=4)
```

```python
tf.keras.models.save_model(model,"Test2.h5")
```
