```python
def dynamic_cell_state(lst,epoch_n=1,lpm=0):
  temp=[]
  for i in range(16):
    m=lst[i]
    mns=[]
    for j in range(lpm,11591,190):
      L=m.loc[j:j+(99*19)].to_numpy().flatten()
      M=m.loc[j+100*19:j+109*19].to_numpy().flatten()
      mns.append([L,M])
    mns=np.array(mns)
    mns=mns[0:71]
    model=tf.keras.Sequential([tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(512,return_sequences=True),input_shape=(300,1)),
                    tf.keras.layers.Dense(128),
                    tf.keras.layers.Dropout(0.1),
                    tf.keras.layers.GRU(units=32),
                    tf.keras.layers.Dropout(0.1),
                    tf.keras.layers.Dense(32),
                    tf.keras.layers.Dense(30)])
    model.compile(loss='mean_squared_error',optimizer='Adam',metrics=['mean_absolute_error'])
    for jk in range(0,50):
      model.fit(np.array([mns[jk][0]]),np.array([mns[jk][1]]),epochs=epoch_n,verbose=0)
    temp1=lst[i].copy()
    temp1=temp1.to_numpy()
    temp1=pd.DataFrame(temp1)
    for jkl in range(700,811,10):
      m=model.predict(np.array([temp1.loc[(jkl-100):(jkl-1),[0, 1, 2]].to_numpy().flatten()]))
      m=pd.DataFrame(m.reshape((10,3)))
      temp1=pd.concat([temp1,m],ignore_index=True)
    temp.append(temp1)
  return temp
```
